# Entanglement & Bell States


Bell states are recognised as maximum entangled states .

There are 4 bell states:<br/>
a) |&phi;<sup>+</sup>$\rangle$= $\frac{1}{√2}$(|00$\rangle$+|11$\rangle$)<br/>
b) |&phi;<sup>-</sup>$\rangle$= $\frac{1}{√2}$(|00$\rangle$ -|11$\rangle$)<br/>
c) |&psi;<sup>+</sup>$\rangle$= $\frac{1}{√2}$(|01$\rangle$+|10$\rangle$)<br/>
d) |&psi;<sup>-</sup>$\rangle$= $\frac{1}{√2}$(|01$\rangle$-|10$\rangle$)<br/> 

There are multiqubits states also those follow Bell theorem like GHZ state ,W state etc.

There is one more multiqubit state (3-qubit) which is maximum entangled like bell states in Quantum Mechanics i.e Greenberger–Horne–Zeilinger state (GHZ).</br>
|GHZ$\rangle$= $\frac{1}{√2}$(|000$\rangle$+|111$\rangle$)<br/>

```python
qc5=QuantumCircuit(3,3)
qc5.h(0)
qc5.cx(0,1)
qc5.cx(1,2)
#qc5.measure_all()
qc5.draw(output="mpl")
```

In general M-qubit GHZ can represented as:<br/>
|GHZ$\rangle$= $\frac{1}{√2}$(|00000..$\rangle$+|11111..$\rangle$)

```python
#Method to make M-qubit GHZ state.
def GHZ(m):
    if m==1:
            print("You need atleast 2-qubits for GHZ state.")
    else:
        qc=QuantumCircuit(m)
        qc.h(0)
        for k in range(m-1):
            qc.cx(k,k+1)
        qc.measure_all()
        return qc
```

```python
GHZ(5).draw(output="mpl")
```
